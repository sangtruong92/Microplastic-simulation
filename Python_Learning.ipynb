````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as gaussian_filter
import warnings
warnings.filterwarnings("ignore")

print("libraries imported successfully")
````

Step 1: Define Camera and System Parameters


````python
# Camera setting
Image_Width = 1920      # pixels
Image_Height = 1000     # pixels
Pixel_Size_X = 1.45     # um/pixel

# Optical system
Focal_Length = 6.0      # lens focal length in mm
Water_Depth_mm = 5.0    # water cotainer depth
n_water =1.33          # refractive index of water
n_air = 1.0           # refractive index of air

# Particle parameters
Particle_Size_Min_um = 50  # minimum particle size in um
Particle_Size_Max_um = 500 # maximum particle size in um
Number_particles = 20

# parameters put into {}
print(f"Image size: {Image_Width} x {Image_Height} pixels")
print(f"Pixel size: {Pixel_Size_X} um/pixel")
print(f"Simulating {Number_particles} particles")
````

Step 2: Create Particle Generation function

````python
def create_particles(center_x, center_y, size_px, image_shape):
    h, w = image_shape
    y, x = np.ogrid[:h, :w] # ogrid mean open grid, which creates two arrays: one for y-coordinates and one for x-coordinates.
    radius = size_px /2.0
    
    # create circular mask 
    distance = np.sqrt((x-center_x)**2 + (y-center_y)**2)
    mask = (distance <= radius).astype(float)
    
    return mask 
    
def micrometers_to_pixels(size_um, focal_length_mm, depth_mm):
    
    # Simplified magnification calculation
    apparent_depth = depth_mm / N_Water # refraction effect
    magnification = focal_length_mm / (50 + apparent_depth)
    
    # Convert to image size
    image_size_m = size_um * 1e-6 * magnification 
    size_pixels = image_size_m / (Pixel_size_um * 1e-6)
    
    return size_pixels

print ("particle generation fuction defined defined!")
    
````

````python
def apply_optical_effects(particle_mask, depth_mm, focal_plane_mm=2.5):
    # calculate defocus
    defocus = abs(depth_mm - focal_plane_mm)
    
    # blur amount increase with defocus
    blur_sigma = 0.5 + 3.0 *defocus
    
    # Apply Gaussian blur (simulates point )
    blurred = gaussian_filter(particle_mask, sigma=blur_sigma)
    
    # Reduce intensity for out-of-focus particles
    intensity_factor = 1.0 / (1.0 + defocus**2)
    blurred = blurred * intensity_factor
    
    return blurred

print("Optical effect function defined")
    
````

Step 4: Main Simulation function

````python
def simulate_microplastic_image(num_particles=20):
    
    image = np.zeros ((), dtype=np.float64)
    
    particles_info = []
    
    print(f"Generating {num_particles} particles ... ")
    
    for i in range(num_particles):
        # random particle properties
        true_size_um = np.random.uniform(Particle_Size_Min_um, Particle_Size_Max_um)
        depth_mm = np.random.unform(0, Water_Depth_mm)
        
        # Random position 
        center_x = np.random.randint(50, Image_Width - 50)
        center_y = np.random.randint(50, Image_Height - 50)
        
        # Convert size to pixels 
        size_px = micrometers_to_pixels(true_size_um, focal_length_mm, depth_mm)
        
        # Create particle 
        particle = create_particles(center_x, center_y, size_px, image.shape)
        
        # Apply optical effects
        particle = apply_optical_effects(particle, depth_mm, focal_plane_mm=FOCAL_PLANE_DEPTH_MM)
        
    
    return image, particles_info

print(" Main simulation function ready!")
````

## 📚 Code Explanation: List Comprehension & f-string

Let's break down this complex line:
```python
print(f"Average particle size: { np.mean( [ p['size_um']for p in particles]):.1f }  μm")
```

````python
# Step-by-step breakdown of: np.mean([p['size_um'] for p in particles])

print("="*70)
print("BREAKING DOWN THE CODE")
print("="*70)

# Example data structure (like what the simulation creates)
particles = [
    {'id': 1, 'size_um': 120.5, 'depth_mm': 1.2},
    {'id': 2, 'size_um': 250.3, 'depth_mm': 2.5},
    {'id': 3, 'size_um': 180.7, 'depth_mm': 3.1},
    {'id': 4, 'size_um': 320.9, 'depth_mm': 1.8},
    {'id': 5, 'size_um': 200.1, 'depth_mm': 4.2}
]

print("\nExample particles data:")
for p in particles:
    print(f"  {p}")

print("\n" + "="*70)
print("STEP 1: Understanding 'particles' - List of Dictionaries")
print("="*70)
print("particles is a LIST containing DICTIONARIES")
print(f"Number of particles: {len(particles)}")
print(f"Type: {type(particles)}")
print(f"First particle: {particles[0]}")
print(f"Type of first particle: {type(particles[0])}")

print("\n" + "="*70)
print("STEP 2: Dictionary Access - p['size_um']")
print("="*70)
print("Each particle 'p' is a dictionary with keys: 'id', 'size_um', 'depth_mm'")
print("p['size_um'] gets the size value from that particle")
print("\nExamples:")
for i, p in enumerate(particles[:3]):
    print(f"  particles[{i}]['size_um'] = {p['size_um']} μm")

print("\n" + "="*70)
print("STEP 3: List Comprehension - [p['size_um'] for p in particles]")
print("="*70)
print("This creates a NEW LIST containing ONLY the sizes")
print("\nLong way (using a loop):")
sizes_long = []
for p in particles:
    sizes_long.append(p['size_um'])
print(f"  sizes = {sizes_long}")

print("\nShort way (list comprehension):")
sizes_short = [p['size_um'] for p in particles]
print(f"  sizes = {sizes_short}")

print(f"\n✓ Both methods give the same result: {sizes_long == sizes_short}")

print("\n" + "="*70)
print("STEP 4: NumPy Mean - np.mean([...])") 
print("="*70)
sizes = [p['size_um'] for p in particles]
average = np.mean(sizes)
print(f"Input list: {sizes}")
print(f"Sum: {sum(sizes)}")
print(f"Count: {len(sizes)}")
print(f"Average: {sum(sizes)} / {len(sizes)} = {average}")
print(f"np.mean(sizes) = {np.mean(sizes)}")

print("\n" + "="*70)
print("STEP 5: f-string Formatting - :.1f")
print("="*70)
print("The ':.1f' formats the number:")
print("  : = start of format specification")
print("  .1 = show 1 decimal place")
print("  f = format as float (decimal number)")
print("\nExamples:")
print(f"  No formatting: {average}")
print(f"  With :.1f: {average:.1f}")
print(f"  With :.2f: {average:.2f}")
print(f"  With :.0f: {average:.0f}")

print("\n" + "="*70)
print("STEP 6: Complete f-string - f\"Average particle size: {...}\"")
print("="*70)
print("f-strings allow you to embed expressions inside { }")
result = f"Average particle size: {np.mean([p['size_um'] for p in particles]):.1f} μm"
print(f"Output: {result}")

print("\n" + "="*70)
print("VISUAL BREAKDOWN")
print("="*70)
print("""
f"Average particle size: { np.mean( [ p['size_um']for p in particles]):.1f }  μm"
│                          │       │ │          │                      │││
│                          │       │ │          │                      │││
│                          │       │ │          │                      ││└─── f = float format
│                          │       │ │          │                      │└──── .1 = one decimal
│                          │       │ │          │                      └───── : = format specifier
│                          │       │ │          └──────────────────────────── loop through particles
│                          │       │ └─────────────────────────────────────── extract 'size_um' value
│                          │       └───────────────────────────────────────── list comprehension
│                          └───────────────────────────────────────────────── calculate mean
└──────────────────────────────────────────────────────────────────────────── f-string prefix
""")
````

### 🎯 Summary

**What the code does:**
1. **Loops** through all particles
2. **Extracts** each particle's size (`size_um`)
3. **Calculates** the average of all sizes
4. **Formats** the number to 1 decimal place
5. **Prints** a nice message

**Equivalent longer version:**

- **Format specification**: `:.1f`
